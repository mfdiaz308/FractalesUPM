### an_grafico

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_aux(interv,f):
  if f==0: # lineal
    A = -0.5
    B = -1
    x1 = np.arange(interv[0],interv[1],0.1)
    y1 = A*x1 + B
  if f==1: # logistica
    C = 3.55
    x1 = np.arange(interv[0],interv[1],0.01)
    y1 = C*x1*(1-x1)
  if f==2: # cuadratica
    C = 0.5
    x1 = np.arange(interv[0],interv[1],0.1)
    y1 = x1**2+C
  plt.plot(x1,y1)

  # 45ยบ
  x = np.arange(interv[0],interv[1],0.01)
  y = x
  plt.plot(x,y,linestyle='dashed',color='black')


def func_graf(x,n):
  if n==0: # lineal
    A = -0.5
    B = -1
    return A*x+B
  if n==1: # logistica
    C = 3.55
    return C*x*(1-x)
  if n==2: # cuadratica
    C = 0.5
    return x**2+C

def vert(id,n):
    id2=[0,0]
    id2[0] = id[0]
    if n==0: # lineal
        id2[1] = func_graf(id[0],n)
    elif n==1: # logistica
        id2[1] = func_graf(id[0],n)
    else: # cuadratica
        id2[1] = func_graf(id[0],n)

    plt.vlines(id2[0],id[1],id2[1],'r')

    return id2

def horiz(id):
    id2 = [0,0]
    id2[0] = id[1]
    id2[1] = id[1]

    plt.hlines(id2[1],id2[0],id[0],'r')

    return id2

def an_grafico(f,p,n,cont):
    if n>0:
        if cont%2==0:  # Par -> horiz
            an_grafico(f,horiz(p),n-1,cont+1)
        else:  # Impar -> vert
            an_grafico(f,vert(p,f),n-1,cont+1)

    plt.show()            
```

```python
plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

# plot_aux(interv,f)
plot_aux([-3,4],0)

# an_grafico(f,p,n,cont)
an_grafico(0,[3,0],10,cont=1) # lineal
```

```python
import matplotlib.pyplot as plt
import numpy as np

def func(p,n):
  if n==0: # lineal
    A = -0.75
    B = 1
    return [A*p[0]+B]
  if n==1: # logistica
    C = 3.55
    return [C*p[0]*(1-p[0])]
  if n==2: # cuadratica
    C = 0.5
    return [p[0]**2+C]

def orb_num(f,p,n):
  if n>0:
    print(p)
    p2 = func(p,f)
    orb_num(f,p2,n-1)


def plot_aux(interv,f):
  if f==0: # lineal
    A = -0.75
    B = 1
    x1 = np.arange(interv[0],interv[1],0.1)
    y1 = A*x1 + B
  if f==1: # logistica
    C = 3.55
    x1 = np.arange(interv[0],interv[1],0.01)
    y1 = C*x1*(1-x1)
  if f==2: # cuadratica
    C = 0.5
    x1 = np.arange(interv[0],interv[1],0.1)
    y1 = x1**2+C
  plt.plot(x1,y1)

  # 45ยบ
  x = np.arange(interv[0],interv[1],0.01)
  y = x
  plt.plot(x,y,linestyle='dashed',color='black')


def func_graf(x,n):
  if n==0: # lineal
    A = -0.75
    B = 1
    return A*x+B
  if n==1: # logistica
    C = 3.55
    return C*x*(1-x)
  if n==2: # cuadratica
    C = 0.5
    return x**2+C

def vert(id,n):
    id2=[0,0]
    id2[0] = id[0]
    if n==0: # lineal
        id2[1] = func_graf(id[0],n)
    elif n==1: # logistica
        id2[1] = func_graf(id[0],n)
    else: # cuadratica
        id2[1] = func_graf(id[0],n)

    plt.vlines(id2[0],id[1],id2[1],'r')

    return id2

def horiz(id):
    id2 = [0,0]
    id2[0] = id[1]
    id2[1] = id[1]

    plt.hlines(id2[1],id2[0],id[0],'r')

    return id2

def an_grafico(f,p,n,cont):
    if n>0:
        if cont%2==0:  # Par -> horiz
            an_grafico(f,horiz(p),n-1,cont+1)
        else:  # Impar -> vert
            an_grafico(f,vert(p,f),n-1,cont+1)

    plt.show()    
```

```python
p = [0.9,0]

orb_num(1,[p[0]],20)

plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

plot_aux([0,1],1)

an_grafico(1,p,20,cont=1)
```

```python
p = [0,0]

orb_num(1,[p[0]],20)

plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

plot_aux([0,1],1)

an_grafico(1,p,20,cont=1)
```

```python
p = [0.1,0]

orb_num(1,[p[0]],20)

plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

plot_aux([0,1],1)

an_grafico(1,p,20,cont=1)
```

```python
p = [0.5,0]

orb_num(1,[p[0]],20)

plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

plot_aux([0,1],1)

an_grafico(1,p,20,cont=1)
```

```python
def orb_num_2d(p,n):
  if n>0:
    print(p)
    orb_num_2d(func(p),n-1)
```

```python
def orb_num_2d_av(p,n,m,cont):
  if n>0:
    if cont>=m:
      print(p)
    orb_num_2d_av(func(p),n-1,m,cont+1)
```

```python
import matplotlib.pyplot as plt

def an_graf_2d(p,n,cont,interv):
  if n>0:
    if cont==0:
      plt.vlines(0,interv[0],interv[1],'k')
      plt.hlines(0,interv[0],interv[1],'k')
      plt.plot(p[0],p[1],'bs')
    else:
      plt.plot(p[0],p[1],'b+')
    an_graf_2d(func(p),n-1,cont+1,interv)
  plt.show()
```

```python
def an_graf_2d_av(p,n,m,cont,interv):
  if n>0:
    if cont==0:
      plt.vlines(0,interv[0],interv[1],'k')
      plt.hlines(0,interv[0],interv[1],'k')
      plt.plot(p[0],p[1],'bs')
    elif cont>=m:
      plt.plot(p[0],p[1],'b+')
    an_graf_2d_av(func(p),n-1,m,cont+1,interv)
  plt.show()
```

```python
def an_graf_2d_lineas(p,n,cont,colors,interv):
    if n>0:
        if cont==0:
            plt.vlines(0,interv[0],interv[1],'k')
            plt.hlines(0,interv[0],interv[1],'k')
            plt.plot(p[0],p[1],'s',color='blue')
        else:
            plt.plot(p[0],p[1],'+',color='blue')
        p2 = func(p)
        plt.plot([p[0],p2[0]],[p[1],p2[1]],color=colors)
        plt.plot(p2[0],p2[1],'+',color='blue')
        an_graf_2d_lineas(p2,n-1,cont+1,colors,interv)
```

```python
def an_graf_2d_lineas_av(p,n,m,cont,colors,interv):
    if n>0:
      if cont>m:
        if cont==0:
            plt.vlines(0,interv[0],interv[1],'k')
            plt.hlines(0,interv[0],interv[1],'k')
            plt.plot(p[0],p[1],'s',color='blue')
        else:
            plt.plot(p[0],p[1],'+',color='blue')
        p2 = func(p)
        plt.plot([p[0],p2[0]],[p[1],p2[1]],color=colors)
        plt.plot(p2[0],p2[1],'+',color='blue')
      an_graf_2d_lineas_av(func(p),n-1,m,cont+1,colors,interv)
```

```python
def an_graf_2d_lineas_animado(p,n,cont,colors,interv):
    for i in range(n+1):
        ax = plt.subplot(111)
        ax.set_aspect(1)

        plt.vlines(0,interv[0],interv[1],'k')
        plt.hlines(0,interv[0],interv[1],'k')

        an_graf_2d_lineas(p,i,0,colors,interv)

        plt.draw()
        plt.pause(0.2)
        plt.clf()

    plt.figure
    ax = plt.subplot(111)
    ax.set_aspect(1)

    an_graf_2d_lineas(p,n,0,colors,interv)

    plt.show()
```
