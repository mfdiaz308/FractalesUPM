## Ejercicio 5

```python
import numpy as np
import matplotlib.pyplot as plt

def feigenLog(a, b, m, n, x, r):
  l = []
  for i in range(r+1):
    s = x
    c = a+i/r*(b-a)
    for j in range(m):
        s = c*s*(1 - s)
    for k in range(n):
        s = c*s*(1 - s)
        l.append([c, s])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')

  plt.show()
```

```python
feigenLog(0,4,100,50,1/2,10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def feigenCuad(a, b, m, n, x, r):
  l = []
  for i in range(r+1):
    s = x
    c = a+i/r*(b-a)
    for j in range(m):
        s = s**2+c
    for k in range(n):
        s = s**2+c
        l.append([c, s])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')

  plt.show()
```

## Ejercicio 7

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def feigenCos(a, b, m, n, x, r):
  l = []
  for i in range(r+1):
    s = x
    c = a+i/r*(b-a)
    for j in range(m):
        s = math.cos(s*c)
    for k in range(n):
        s = math.cos(s*c)
        l.append([c, s])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')

  plt.show()
```

```python
feigenCos(-3,3,100,50,0.5,20000)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def func(x,y,n):
  if n==1:
    return [5/12*(x-np.sqrt(3)*y+x*2*y), 5/12*(np.sqrt(3)*x+y+y*2*x)]
  elif n==2:
    return [0.5*x+2*x*y,-2/3*y+x**2]
  elif n==3:
    return [1+y-7/5*x**2,3/10*x]
  else:
    return [x**2-y**2-3/20,2*x*y+3/4]

def cuenca(a,b,c,d,p,r,n,aux):
  l=[]
  p2 = p
  for i in range(n+1):
    p2 = func(p2[0],p2[1],aux)
  for i in range(r+1):
    s = [a,c]
    s[0] = a + i/r * (b-a)
    for j in range(r+1):
      s[1] = c + j/r * (d-c)
      s2 = func(s[0],s[1],aux)
      for k in range(n):
        s2 = func(s2[0],s2[1],aux)
        if math.sqrt((s2[0]-p2[0])**2 + (s2[1]-p2[1])**2) < 10**(-6):
          l.append([s[0],s[1]])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k.')
  plt.show()
```

```python
cuenca(-1,1,-1,0.5,[-0.5,-0.5],400,60,1)
```

```python
cuenca(-0.5,0.5,-0.5,0.5,[0,0],600,25,2)
```

```python
cuenca(-1,1,-1,1,[1,1],500,8,3)
```
