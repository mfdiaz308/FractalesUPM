## Conjuntos de Julia

```python
import matplotlib.pyplot as plt
import numpy as np
import math

def func(z,c,n):
  if n==0: # Cuadratica
    return z*z+c
  if n==1: # Logistica
    return c*z*(1-z)
  if n==2: # Cubica
    return z*z*z+c


def juliaCuad(c,aRect,bRect,cRect,dRect,r,n):
  l = []
  for i in range(r+1):
    for j in range(r+1):
      s = complex(aRect + i/r * (bRect-aRect),cRect + j/r * (dRect-cRect))
      s2 = s
      for k in range(n+1):
          s2 = func(s2, c, 0)
      if math.sqrt(s2.real**2+s2.imag**2) < 2:
          l.append([s.real, s.imag])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')
  plt.show()

def juliaLog(c,aRect,bRect,cRect,dRect,r,n):
  l = []
  for i in range(r+1):
    for j in range(r+1):
      s = complex(aRect + i/r * (bRect-aRect),cRect + j/r * (dRect-cRect))
      s2 = s
      for k in range(n+1):
          s2 = func(s2, c, 1)
      if math.sqrt(s2.real**2+s2.imag**2) < 2:
          l.append([s.real, s.imag])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')
  plt.show()

def juliaCub(c,aRect,bRect,cRect,dRect,r,n):
  l = []
  for i in range(r+1):
    for j in range(r+1):
      s = complex(aRect + i/r * (bRect-aRect),cRect + j/r * (dRect-cRect))
      s2 = s
      for k in range(n+1):
          s2 = func(s2, c, 2)
      if math.sqrt(s2.real**2+s2.imag**2) < 2:
          l.append([s.real, s.imag])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')
  plt.show()
```

#### Julia cuadratico

```python
plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

juliaCuad(complex(-1,0),-2,2,-1.5,1.5,1000,100)
```

#### Julia logistico

```python
juliaLog(complex(0,1),-1.5,2.5,-2.5,2.5,1000,200)
```

#### Julia cubico

```python
juliaCub(complex(0.57,-0.5),-1.5,2.5,-2.5,2.5,1000,100)
```

## Conjuntos de Mandelbrot

```python
import cmath

def func(z,c,n):
  if n==0: # Cuadratica
    return z*z+c
  if n==1: # Logistica
    return c*z*(1-z)
  if n==2: # Cubica
    return z*z*z+c

def mandCuad(aRect,bRect,cRect,dRect,r,n):
  cList = []
  for i in range(r+1):
    for j in range(r+1):
      c2 = complex(aRect + i/r * (bRect-aRect),cRect + j/r * (dRect-cRect))
      cList.append(c2)
  l = []
  for c in cList:
    s2 = complex(0,0)
    for k in range(n+1):
        s2 = func(s2, c, 0)
    if abs(s2) < 2:
        l.append([c.real, c.imag])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')
  plt.show()

def mandLog(aRect,bRect,cRect,dRect,r,n):
  cList = []
  for i in range(r+1):
    for j in range(r+1):
      c2 = complex(aRect + i/r * (bRect-aRect),cRect + j/r * (dRect-cRect))
      cList.append(c2)
  l = []
  for c in cList:
    s2 = complex(0.5,0)
    for k in range(n+1):
        s2 = func(s2, c, 1)
    if abs(s2) < 2:
        l.append([c.real, c.imag])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')
  plt.show()

def mandCub(aRect,bRect,cRect,dRect,r,n):
  cList = []
  for i in range(r+1):
    for j in range(r+1):
      c2 = complex(aRect + i/r * (bRect-aRect),cRect + j/r * (dRect-cRect))
      cList.append(c2)
  l = []
  for c in cList:
    s2 = complex(0,0)
    for k in range(n+1):
        s2 = func(s2, c, 2)
    if abs(s2) < 2:
        l.append([c.real, c.imag])

  plt.plot([i[0] for i in l], [i[1] for i in l],'k,')
  plt.show()
```

#### Mandelbrot función cuadrática

```python
plt.figure
ax = plt.subplot(111)
ax.set_aspect(1)

mandCuad(-2,2,-2,2,1000,200)
```
