# SFI determinista

```python
import matplotlib.pyplot as plt
import numpy as np

def sfi(m,n,p):
    if n>0:
        pAux = []
        for j in range(0,len(m)):
            pAux = np.array([(m[j][0]*p[0] + m[j][1]*p[1] + m[j][4]),
                             (m[j][2]*p[0] + m[j][3]*p[1] + m[j][5])])
            sfi(m,n-1,pAux)
    else:
        plt.plot(p[0],p[1],marker='.',color='black')

def exec(m,N_ITER):
  p=np.array([0,0])

  plt.figure()
  ax = plt.subplot(111)
  ax.set_aspect(1)
  sfi(m,N_ITER,p)

  plt.show()
```

# SFI aleatorio

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def sfi_al(m):
  N_ITER = 10000
  pAux = np.array([0,0])

  for i in range(N_ITER):
      prob = np.array([])
      dado = random.randint(0,100000)

      for j in range(len(m)):
          prob = np.append(prob,m[j][6] * 100000)

      aux = 0
      for j in range(0,len(prob)):
          if j==0: # primera iteracion
              if 0 < dado <= prob[j]:
                  break
          elif j==len(prob)-1: # ultima iteracion
              break
          else:
              if aux < dado <= aux+prob[j]:
                  break
          aux += prob[j]

      pAux = np.array([(m[j][0]*pAux[0] + m[j][1]*pAux[1] + m[j][4]),
                      (m[j][2]*pAux[0] + m[j][3]*pAux[1] + m[j][5])])
      if i>=50:
          plt.plot(pAux[0],pAux[1],marker='.',color='black')

  ax = plt.subplot(111)
  ax.set_aspect(1)

  plt.show()
```

```python
# Ej 1 transparencias
m = np.array([
    [0.5,0,0,0.5,0,0,0.333],
    [0.5,0,0,0.5,0.5,0,0.333],
    [0.5,0,0,0.5,0.25,0.433,0.334]
])

sfi_al(m)
```
